```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

results_file = Path("../training_results/first_run2/results.csv")

if results_file.exists():
    results_df = pd.read_csv(results_file)
    
    # Plot the metrics
    plt.figure(figsize=(15, 10))
    
    # Box Loss
    plt.subplot(2, 3, 1)
    plt.plot(results_df['train/box_loss'], label='Train')
    plt.plot(results_df['val/box_loss'], label='Val')
    plt.title('Box Loss')
    plt.legend()
    
    # Class Loss
    plt.subplot(2, 3, 2)
    plt.plot(results_df['train/cls_loss'], label='Train')
    plt.plot(results_df['val/cls_loss'], label='Val')
    plt.title('Class Loss')
    plt.legend()
    
    # DFL Loss
    plt.subplot(2, 3, 3)
    plt.plot(results_df['train/dfl_loss'], label='Train')
    plt.plot(results_df['val/dfl_loss'], label='Val')
    plt.title('DFL Loss')
    plt.legend()
    
    # Precision, Recall, mAP
    plt.subplot(2, 3, 4)
    plt.plot(results_df['metrics/precision(B)'], label='Precision')
    plt.plot(results_df['metrics/recall(B)'], label='Recall')
    plt.title('Precision & Recall')
    plt.legend()
    
    # mAP metrics
    plt.subplot(2, 3, 5)
    plt.plot(results_df['metrics/mAP50(B)'], label='mAP@0.5')
    plt.plot(results_df['metrics/mAP50-95(B)'], label='mAP@0.5:0.95')
    plt.title('mAP Metrics')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('training_metrics.png')
    plt.show()
    
    print("Final mAP@0.5:", results_df['metrics/mAP50(B)'].iloc[-1])
else:
    print(f"Results file not found at {results_file}")
```
